# Creating plots using object-oriented version

```python
# first we will create a figure, then we will add an axis
# 1st method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
fig1 = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
```

```python
# 3rd method - recommended
x = [1,2,3,4]
y = [11,22,33,44]
fig,ax = plt.subplots()
ax.plots(x,y);
```

# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
#1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]
# 2.Setup plot
fig,ax = plt.subplots(figsize = (10,10)) #width,height
#3. plot the data
ax.plot(x,y)
#4. Customize plot
ax.set(title = "Simple plot",xlabel = "x-axis",ylabel = "y-axis")
#5. Save and show(you save the whole figure)
fig.savefig("sample_plot.png")
```

# Plotting data from pandas dataframes

```python
import pandas as pd
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# Line plot from pandas DataFrame
# Start with some dummy data
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2024', periods=1000))

# Note: ts = short for time series (data over time)

ts = ts.cumsum()
ts
ts.plot()
```

```python
# Working with actual data
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '',regex=True) # Tell pandas to replace using regex
car_sales
```

```python
# removing last 2 zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale date"] = pd.date_range("1/1/2024",periods = len(car_sales))
car_sales
```

```python
car_sales["Total sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
#plotting the total sales data
car_sales.plot(x = "Sale date", y = "Total sales")
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x = "Odometer (KM)", y = "Price", kind = "scatter");
```

```python
over_5k = car_sales[car_sales["Price"]>6500]
over_5k
```

```python
# Using Pyplot method
over_5k.plot(kind = "scatter", x = "Make",y = "Price")
```

```python
# Using XX method
```

```python
fig,ax = plt.subplots(figsize = (10,6))
over_5k.plot(kind = "scatter", x= "Make", y = "Price",ax=ax);
```

```python
ax.set_ylim([10000,30000])
```

```python
# XX method from scratch
fig,ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(x = over_5k["Make"], y = over_5k["Odometer (KM)"])
#customize the plot
ax.set(title = "Car sales", xlabel = "Brand", ylabel = "KM");
```
